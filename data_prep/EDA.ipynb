Разделим строчку с жанрами на разные колонки и посмотрим что вообще есть в них

```python
split_set = df['Genres'].str.split(',', n = 13, expand = True)
split_set
```

```python
split_set.isna().sum() 
```

```python
split_set[0].unique()
```

```python
(split_set == 'Hentai').sum()
```

```python
ind2drop = split_set.loc[(split_set[0] == 'Hentai')|(split_set[1] == ' Hentai')|(split_set[2] == ' Hentai')|(split_set[3] == ' Hentai')].index
```

```python
split_set.drop(index = ind2drop, inplace = True)
```

```python
df.drop(index = ind2drop, inplace = True)
```

По табличке выше можно понять что за очень редким исключением аниме не имеют более 5 жанров. Давайте оставим их а остальными позволим себе принебречь

```python
df.drop(columns = ['Genres'], inplace = True)
df['Genre_1'] = split_set[0]
df['Genre_2'] = split_set[1]
df['Genre_3'] = split_set[2]
df['Genre_4'] = split_set[3]
df['Genre_5'] = split_set[4]
```

Судя по всему это либо какие то очень малоизвестные аниме либо не увидевшие свет премьеры. Давайте избавимся от таких объектов.
