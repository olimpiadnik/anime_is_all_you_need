Пока мы ни разу не смотрели на жанры аниме, давайте посмотрим насколько часто встречаются различные типы и как они влияют на популярность аниме

```python
from collections import Counter

anime.fillna('nan', inplace = True)
anime.loc[:, ['Genre_1']] = anime['Genre_1'].apply(lambda x: x.strip())
anime.loc[:, ['Genre_2']] = anime['Genre_2'].apply(lambda x: x.strip())
anime.loc[:, ['Genre_3']] = anime['Genre_3'].apply(lambda x: x.strip())
anime.loc[:, ['Genre_4']] = anime['Genre_4'].apply(lambda x: x.strip())
anime.loc[:, ['Genre_5']] = anime['Genre_5'].apply(lambda x: x.strip())
genres = pd.concat([anime['Genre_1'], anime['Genre_2'], anime['Genre_3'], anime['Genre_4'], anime['Genre_5']], ignore_index = True)
genres.dropna(inplace = True)

from collections import Counter

c = Counter(genres)
print(c.most_common()[1:])

top_genres = ['Action', 'Comedy', 'Adventure', 'Fantasy', 'Sci-Fi']

plt.figure(figsize = (15, 6))
sns.boxplot(anime.loc[anime['Genre_1'].isin(top_genres)],
            x = 'Genre_1',
            y = 'Favorites');
```

В целом, все ожидаемо, экшн, комедия и приключения - три столпа аниме. Давайте возьмем первые 5 самых популярных жанра и построим по ним коробочки с усами для количества добавлений в папку favourites

Легко понять, что среди данных очень много нулей или близких к нулю вхождений

## Генерация фичей

Мне интересно попробовать повытаскивать дополнительные фичи для каждого объекта, давайте попробуем придумать что-то интересное

```python
# для начала откатимся к неотнормированой версии
anime = df[df['Type'] == 'TV'].copy(deep = True)
anime.drop(columns = ['Type'], inplace = True)
anime = anime.loc[(anime['Favorites'] < 75000)&(anime['Plan to Watch'] < 300000)&(anime['Members'] < 1500000)&(anime['Completed'] < 1500000)&(anime['Watching'] < 200000)&(anime['Dropped'] < 100000)]
anime
```

```python
anime['Finish_conv'] = anime['Completed']/anime['Members'] # доля людей досмотревших аниме (стоит отметит что я считаю что все кто включен в members начинали смотреть аниме)
anime['Drop_rate'] = anime['Dropped']/anime['Members']
anime['Fav_conv'] = anime['Favorites']/anime['Members']
anime['Love_rate'] = (anime['Score-10']+anime['Score-9'])/(anime['Score-1']+anime['Score-2']+anime['Score-3']+anime['Score-4']+anime['Score-5']+anime['Score-6']+anime['Score-7']+anime['Score-8']+anime['Score-9']+anime['Score-10'])
anime['Hate_rate'] = (anime['Score-1']+anime['Score-2'])/(anime['Score-1']+anime['Score-2']+anime['Score-3']+anime['Score-4']+anime['Score-5']+anime['Score-6']+anime['Score-7']+anime['Score-8']+anime['Score-9']+anime['Score-10'])
anime['comp_dropped'] = anime['Completed']/anime['Dropped']
```

```python
anime_dum = pd.get_dummies(anime.drop(columns = ['Episodes', 'Aired', 'Genre_4', 'Genre_5']))
anime_dum
```

```python
anime_dum.to_csv('anime_dum.csv')
```
